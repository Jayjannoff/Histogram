```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-white')
x = np.random.randn(1000)
plt.hist(x,bins=30,alpha=0.5,histtype='stepfilled',color='steelblue',edgecolor='red')
```

```python
x1 = np.random.normal(0,0.8,1001)
x2 = np.random.normal(-2,1,1001)
x3 = np.random.normal(2.5,3,1001)

info = dict(histtype='stepfilled',alpha=0.4,bins=40)
```

```python
plt.hist(x1,**info)
plt.hist(x2,**info)
plt.hist(x3,**info)
```

```python
mean = [0,0]
covy = [[1,1],[1,3]]
x,y = np.random.multivariate_normal(mean, covy,10000).T
```

```python
plt.hist2d(x,y,cmap='Reds',bins=30)
a = plt.colorbar()
a.set_label('the color which shows the height of other things')
```

```python
boy = [2,6]
#the boy shows where the main grid place
kenglik = [[2,2],[2,4]]
#kenglik shows placing of all coloured grid not main
x,y = np.random.multivariate_normal(boy,kenglik,10000).T
```

```python
plt.hexbin(x,y,gridsize = 20,cmap='Blues')
a = plt.colorbar()
a.set_label('And now hexbins will be shown in the place of grids ')
```
